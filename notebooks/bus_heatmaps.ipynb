```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
import folium
import numpy as np
```

```python
we_go = pd.read_csv('../data/partner_project_data.csv')
```

```python
new_graphs = we_go
```

```python
new_graphs.OPERATOR.astype(int)
```

```python
new_graphs['sum_hdwy'] = round(new_graphs['HDWY_DEV'])
```

```python
new_graphs['sum_schd'] = round(new_graphs['SCHEDULED_HDWY'])
```

```python
new_graphs['means']=(new_graphs['sum_schd']/new_graphs['sum_hdwy'])
```

```python
new_graphs = new_graphs[np.isfinite(new_graphs['means'])]
```

```python
new_graphs = new_graphs[(new_graphs['means']>5) & (new_graphs['means']<15)]
```

```python
new_graphs['SCHEDULED_TIME'] = pd.to_datetime(new_graphs['SCHEDULED_TIME'])
```

```python
new_graphs['hour']= new_graphs['SCHEDULED_TIME'].dt.hour
```

```python
new_graphs['minutes']= new_graphs['SCHEDULED_TIME'].dt.minute
```

```python
new_graphs['days']= new_graphs['SCHEDULED_TIME'].dt.day_name()
```

```python
def categorize_hour(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 24:
        return 'Evening'
    else:
        return 'Nightime'
```

```python
new_graphs['ACTUAL_ARRIVAL_TIME'] = pd.to_datetime(new_graphs['ACTUAL_ARRIVAL_TIME'])
```

```python
new_graphs['SCHEDULED_TIME'] = pd.to_datetime(new_graphs['SCHEDULED_TIME'])
```

```python
new_graphs['ARRIVAL_TIME_DIFF'] = new_graphs['ACTUAL_ARRIVAL_TIME'] - new_graphs['SCHEDULED_TIME']
```

new_graphs

```python
new_graphs
```

```python
new_graphs_late = new_graphs[new_graphs['ADJUSTED_LATE_COUNT']==1]
```

```python
new_graphs_late
```

```python
new_graphs_late_filter = new_graphs_late.groupby('ROUTE_ABBR')['OPERATOR'].value_counts().reset_index()
```

```python
new_graphs_late_filter.head()
```

```python
new_graphs_operators_10 = new_graphs_late
```

```python
new_graphs_operators_10
```

```python
filtered_operators = new_graphs_late_filter[new_graphs_late_filter['count'] > 10]['OPERATOR']
```

```python
filtered_main_df = new_graphs[new_graphs['OPERATOR'].isin(filtered_operators)]
```

```python
new_graphs_operators_10 = filtered_main_df.sort_values(by='OPERATOR')
```

```python
new_graphs_operators_10 = new_graphs_operators_10[new_graphs_operators_10['ADJUSTED_LATE_COUNT']==1]
```

```python
new_graphs_operators_10
```

```python
new_graphs_operators_10 = new_graphs_operators_10[(new_graphs_operators_10['means']>5) & (new_graphs_operators_10['means']<15)]
```

```python
new_graphs_operators_10
```

```python
new_graphs_operators_10['time_of_day'] = new_graphs_operators_10['hour'].apply(categorize_hour)
```

```python
new_graphs_operators_10
```

```python
busmap = folium.Map(location=[36.3, -86.7], zoom_start=12, width='80%', height='80%')

# Group by operator
grouped = new_graphs_operators_10.groupby('OPERATOR')

# Define colors for different operators
colors = ['blue', 'green', 'red', 'purple', 'orange']

# Iterate through each group
for i, (operator, group) in enumerate(grouped):
    points = []
    for row_index, row_values in group.iterrows():
        loc = [row_values['LATITUDE'], row_values['LONGITUDE']]
        points.append(loc)  # Add the location to the points list
        pop = f"Operator: {row_values['OPERATOR']}<br>Time Difference:{row_values['ARRIVAL_TIME_DIFF']}<br>Direction:{row_values['ROUTE_DIRECTION_NAME']}<br>Day:{row_values['days']}<br>Time:{row_values['time_of_day']}"
        icon = folium.Icon(color="orange", icon="bus", prefix='fa')
        
        marker = folium.Marker(location=loc, popup=pop, icon=icon)
        marker.add_to(busmap)
    
    # Add the PolyLine to trace the route for each operator
    folium.PolyLine(points, color=colors[i % len(colors)], weight=2.5, opacity=1).add_to(busmap)

# Display the map
busmap
#busmap.save('busmapoutliers_10.html')
```
